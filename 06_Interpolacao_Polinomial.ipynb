# O problema de Interpolação

Em várias situações é possível que conheçamos uma função $f: [a, b] \rightarrow \mathbb{R}$ em alguns pontos apenas. Uma pergunta natural é se é possível, de alguma forma razoável, extrapolar essa informação para todo o intervalo de interesse.

Infelizmente esse problema é muito vago. De fato considere o problema de aproximar a função seno conhecendo os seus valores nos nos ângulos notáveis.

| $x$       | $0$ | $\frac{\pi}{4}$      | $\frac{\pi}{3}$      | $\frac{\pi}{2}$ |
| :-------: | --- | -------------------- | -------------------- | --------------- |
| $\sin(x)$ | $0$ | $\frac{\sqrt{2}}{2}$ | $\frac{\sqrt{3}}{2}$ | $1$             |

Veja o gráfico da função abaixo com os pontos tabelados destacados.

```julia
using PyPlot

# Plota os pontos nos ângulos notáveis
angulos = [0, π/4, π/3, π/2]
scatter(angulos, sin(angulos), color="r")

# Plota o gráfico usando vários pontos.
x = linspace(0, π/2, 100)
plot(x, sin(x), linestyle="--", color="k");
```

Nesse caso é claro que além da função cosseno há múltiplas funções que passam pelos pontos vermelhos. O nosso desejo, entrentando, é buscar uma função que seja "bem comportada", que não possua variações bruscas.

Uma das maneiras de controlar a complexidade das funções que podemos usar na aproximação é definir uma classe du funções de trabalho, que sejam bem conhecidas e bem comportadas. Um dos mais fundamentais tipos de funções bem que são fáceis de calcular são as funções polinomiais.

### Definição: Interpolação Polinomial

Dado um conjunto de pares $(x_i, y_i) \in \mathbb{R} \times \mathbb{R}, i = 1, \ldots, m$, desejamos encontrar um polinômio $p$, tal que $p(x_i) = y_i,\ i = 1, \ldots, m$.
